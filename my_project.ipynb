```python
df["Date"] = pd.to_datetime(df["Date"],format = "%Y-%m-%d")
```

```python
df.dtypes
```

```python
df.rename(columns = {'State/UnionTerritory':"States" , "Cured" : "Recovered"},inplace = True)
```

```python
df.head()
```

```python
df.drop(["Sno" , "Time" , "ConfirmedIndianNational" , "ConfirmedForeignNational"], axis = 1,inplace = True)
```

```python
df.head()
```

```python
df["Active"] = df["Confirmed"] - df["Recovered"] -df["Deaths"]
```

```python
df.head()
```

```python
df[df.duplicated()]
```

```python
df[df.isna()].index
```

```python
#__ No null values__
ind_Cases = df[df['Date']==df['Date'].max()].copy()
```

```python
ind_Cases
```

```python
ind_Cases.index = ind_Cases["States"]
ind_Cases
```

```python
ind_Cases =ind_Cases.drop(['States','Date'],axis=1)
ind_Cases
```

```python
df =pd.DataFrame(pd.to_numeric(ind_Cases.sum()),dtype=np.int64).transpose()
df
```

```python
df.style.background_gradient(cmap='summer_r',axis=1)
```

```python
ind_Cases.sort_values('Active',ascending =False).style\
        .background_gradient(cmap='BrBG_r',subset=['Confirmed'])\
        .background_gradient(cmap='YlGn_r',subset=['Deaths'])\
        .background_gradient(cmap='YlOrBr',subset=['Recovered'])\
        .background_gradient(cmap='BuPu',subset=['Active'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
ind_Cases.reset_index(level=0, inplace=True)
ind_Cases.head()
```

```python
top_10_Confirmed = ind_Cases.sort_values(by='Confirmed', ascending=False).head(10)
top_10_Confirmed
```

```python
sns.barplot(x='Confirmed', y='States', data=top_10_Confirmed, color='grey')
sns.set(rc={'figure.figsize': (10, 6)})
# Add labels and a title
plt.xlabel('Confirmed Cases')
plt.ylabel('States')
plt.title('Top 10 States with Most Confirmed Cases')
```

```python
top_10_Death = ind_Cases.sort_values(by='Deaths', ascending=False).head(10)
sns.barplot(x='Deaths', y='States', data=top_10_Death , color = "#494949")
sns.set(rc={'figure.figsize': (10, 6)})
plt.title('Top 10 States with most Deaths')
```

```python
top_10_Recovered = ind_Cases.sort_values(by='Recovered', ascending=False).head(10)
sns.barplot(x='Recovered', y='States', data=top_10_Recovered , color = "#898989")
sns.set(rc={'figure.figsize': (10, 6)})
plt.title('Top 10 States with most Recovered Patients')
```
